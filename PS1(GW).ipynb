```python
import matplotlib.pyplot as plt
import numpy as np
from astropy.constants import G , M_sun, c
from astropy import units as u

# setting values
m_1 = m_2 = (25*M_sun).to(u.kg)
d_L = (500*u.Mpc).to(u.m)
M = m_1 + m_2
mu = (m_1*m_2) / (m_1 + m_2)
M_c = ((m_1*m_2)**(3/5)) / ((m_1 + m_2)**(1/5))
f_0 = 20
t_m = ((5 * c.value**5)/(256 * np.pi**(8/3) * G.value**(5/3) * M_c.value**(5/3))) * f_0**((-8)/3)  # because this is the limit till the model will actually work

# setting equations
t = np.linspace(0, t_m, 5000)
f = ((- (256/5) * np.pi**(8/3) * ((G.value**(5/3) * M_c.value**(5/3)) / (c.value**5)) * t) + f_0**((-8)/3))**((-3)/8)

# figure size
plt.figure(figsize=(20, 10))  # width=8 inches, height=5 inches

# plotting
plt.plot(t, f, color='purple', label=f'$f(t)$')
plt.axhline(100, color='orange', linestyle ='--', label=f'revolving 50 times per second')

# axis / titile / legend
plt.xlabel('Time [s]')
plt.ylabel('Frequancy [Hz]')
plt.title('...')
plt.legend(loc="best")

# show plot
plt.show()

# setting the rest of the parameters
F_plus = 0.5
F_x = 0.5
omega = np.pi * f
h_plus = -((4 * G.value * mu) / (c.value**4 * d_L.value)) * (G.value * M * omega)**(2/3) * (np.cos(2*omega*t))
h_x = ((4 * G.value * mu) / (c.value**4 * d_L.value)) * (G.value * M * omega)**(2/3) * (np.sin(2*omega*t))

# setting equations
h = F_plus*h_plus + F_x*h_x

# figure size
plt.figure(figsize=(20, 10))

# plotting
plt.plot(t, h, color='purple', label=f'h(t)')

# axis / titile / legend
plt.xlabel('Time [s]')
plt.ylabel('Strain (h)')
plt.title('...')
plt.legend(loc="best")

# show plot
plt.show()
```
